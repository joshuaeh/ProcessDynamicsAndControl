# Animate Plots in Jupyter Notebook
To animate a plot within a loop in Jupyter Notebook, we will use `display` from `IPython`. To begin, when you import other packages, be sure to import display: `from IPython import display`. Next, define the loop that you want to plot. At the end of each iteration, add the lines `display.clear_output(wait=True)` which tells the IPython display that shows the graphs below the cell to wait for the next figure, then clear the previous figure. Including `plt.pause(0.1)` afterward helps keep these graphs separate from one another.  
  
See the Example below for how to animate plots in jupyter. In addition, there is an example of how to save plots to a subfolder so the pictures may be string together to create a .gif
# Example 1 - Sin and Cosine Graphs
## Import Packages

```python
import numpy as np
import matplotlib.pyplot as plt
from IPython import display
```

## Graph in a loop

```python
x = np.linspace(0, 6 * np.pi, 100)
cos_values = []
sin_values = []
for i in range(len(x)):
    # add the sin and cos of x at index i to the lists
    cos_values.append(np.cos(x[i]))
    sin_values.append(np.sin(x[i]))
    
    # graph up until index i
    plt.plot(x[:i], cos_values[:i], label='Cos')
    plt.plot(x[:i], sin_values[:i], label='Sin')
    plt.xlabel('input')
    plt.ylabel('output')
    
    # lines that help with animation
    display.clear_output(wait=True)  # replace figure with next figure when figure is generated
    plt.pause(.01)  # wait between figures
```

# Example 2 - Save as individual PNG files to make gif
Gifs can be made with a variety of online engines.A gif can me made by saving individual photos each loop.   
  
Here, `plt.savefig()` save a .png of each graph generated within the loop within the subfolder named img

```python
x = np.linspace(0, 10, 100)
power_array = np.linspace(0, 10, 100)
for i in range(len(power_array)):
    y = x**power_array[i]
    
    plt.figure(figsize=(5,5))
    plt.plot(x, y)
    plt.xlabel('input')
    plt.ylabel('output')
    plt.savefig('img/'+str(i)+'.png')
```
